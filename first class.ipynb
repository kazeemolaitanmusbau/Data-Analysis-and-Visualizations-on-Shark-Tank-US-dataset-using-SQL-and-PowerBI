# Bitwise (|, &)
##### is used to combine two or more conditions. Also is return boolean result

- |: will return true the moment  one of the conditions is true, also return true if all the conditions are true. but it flase when all the conditions are false

- &: will return the mement all the conditions are true, but one out of all the conditions is false, it will false result

```python
# |
(a < b) | (c> b) # both conditions are true
```

```python
# |
(a > b) | (c > b) # one of the conditions is false
```

```python
# &
(a < b) & (c> b) # both conditions are true
```

```python
# |
(a < b) & (c> b) # both conditions are true
```

```python
# index -------- forward

name = 'olaitan'

# fetch out first 'a' from the above name

name[   2    ]
```
